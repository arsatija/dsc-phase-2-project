# Exploratory Notebook

```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect('../zippedData/im.db')
cursor = con.cursor()
```

```python
schema_df = pd.read_sql("""

SELECT *
FROM sqlite_master

""", con)

schema_df
```

```python
movie_basics = pd.read_sql("select * from movie_basics", con)
movie_basics
```

```python
movie_ratings = pd.read_sql("select * from movie_ratings", con)
movie_ratings
```

```python
joint_movies = pd.read_sql("""
select movie_basics.*, movie_ratings.averagerating, movie_ratings.numvotes from movie_basics inner join movie_ratings on movie_basics.movie_id = movie_ratings.movie_id
""", con)
joint_movies
```

```python
joint_movies[joint_movies["primary_title"] == "Toy Story 3"]
```

```python
movie_gross = pd.read_csv('../zippedData/bom.movie_gross.csv.gz')
movie_gross
```

```python
titles = movie_gross["title"].to_list()
titles
```

```python
joint_movies[joint_movies["primary_title"].isin(titles)]
```

```python
joint_movies_revenue = joint_movies.merge(movie_gross, how="inner", left_on="primary_title", right_on="title")
joint_movies_revenue
```
